Draw density figures of multivariate Ge-N distributions

```r
# 导入需要的Package
suppressMessages(suppressWarnings(library(tidyverse)))
suppressMessages(suppressWarnings(library(Bessel)))
```

```r
# IG-N分布的图像

# 两个维度x1和x2
x1 <- seq(-5, 5, length.out = 100)
x2 <- seq(-5, 5, length.out = 100)

# 设置参数
d <- 2
mu <- c(0, 0)
sigma <- matrix(c(2, 1.5, 1.5, 2), nrow = 2)
lambda <- 1.5

# 定义密度函数
f <- function(x1, x2) {
  x <- c(x1, x2)
  alpha <- t(x - mu) %*% solve(sigma) %*% (x - mu) * lambda + (lambda - 1)^2
  result <- sqrt(lambda) * exp(lambda - 1) / 2^((d-1)/2) / pi^((d+1)/2) / det(sigma)^(1/2) * besselK(sqrt(alpha), 0, expon = FALSE)
  return(result)
}

# 计算每个点的密度
z <- matrix(nrow = 100, ncol = 100)
for (i in 1:100) {
  for (j in 1:100) {
    z[i, j] <- f(x1[i], x2[j])
  }
}

# 进行画图
persp(x1, x2, z, theta = 30, phi = 30, expand = 0.5, col = "lightblue",
      ltheta = 120, shade = 0.75, ticktype = "detailed",
      xlab = "x1", ylab = "x2", zlab = "Density")
```

```r
# Ga-N分布的图像

# 两个维度x1和x2
x1 <- seq(-5, 5, length.out = 100)
x2 <- seq(-5, 5, length.out = 100)

# 设置参数
d <- 2
mu <- c(0, 0)
sigma <- matrix(c(1, -0.5, -0.5, 1), nrow = 2)
lambda <- 5

# 定义密度函数
f <- function(x1, x2) {
  x <- c(x1, x2)
  phi <- t(x - mu) %*% solve(sigma) %*% (x - mu) / 2
  result <- gamma(lambda + 0.5) * (lambda - 1)^lambda / (2*pi)^(d/2) / det(sigma)^(1/2) / gamma(lambda) / (lambda - 1 + phi)^(lambda + 0.5)
  return(result)
}

# 计算每个点的密度
z <- matrix(nrow = 100, ncol = 100)
for (i in 1:100) {
  for (j in 1:100) {
    z[i, j] <- f(x1[i], x2[j])
  }
}

# 进行画图
persp(x1, x2, z, theta = 30, phi = 30, expand = 0.5, col = "lightblue",
      ltheta = 120, shade = 0.75, ticktype = "detailed",
      xlab = "x1", ylab = "x2", zlab = "Density")
```

```r
# IGa-N分布的图像

# 两个维度x1和x2
x1 <- seq(-5, 5, length.out = 100)
x2 <- seq(-5, 5, length.out = 100)

# 设置参数
d <- 2
mu <- c(0, 0)
sigma <- matrix(c(1, 0, 0, 4), nrow = 2)
lambda <- 0.75

# 定义密度函数
f <- function(x1, x2) {
  x <- c(x1, x2)
  alpha <- 2 * lambda * t(x - mu) %*% solve(sigma) %*% (x - mu)
  result <- 2^(-0.5 * lambda + (5 - 2 * d) / 4) * lambda^(lambda / 2 + 0.25) * besselK(sqrt(alpha), -lambda+0.5, expon = FALSE) / pi^(d/2) / det(sigma)^(1/2) / gamma(lambda) / (t(x - mu) %*% solve(sigma) %*% (x - mu))^(-0.5 * lambda + 0.25)
  return(result)
}

# 计算每个点的密度
z <- matrix(nrow = 100, ncol = 100)
for (i in 1:100) {
  for (j in 1:100) {
    z[i, j] <- f(x1[i], x2[j])
  }
}

# 进行画图
persp(x1, x2, z, theta = 30, phi = 30, expand = 0.5, col = "lightblue",
      ltheta = 120, shade = 0.75, ticktype = "detailed",
      xlab = "x1", ylab = "x2", zlab = "Density")
```
